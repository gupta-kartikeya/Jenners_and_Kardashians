```python
# Load Pandas, numpy and matplotlib
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
# read data
web_search=pd.read_csv("../Downloads/data.csv")
web_search.head()
web_search.info()
```

```python
for i in web_search.columns:
    # check for "<" in each column
    if "<" in web_search[i].to_string():
        #remove "<" from the object value
        web_search[i]=web_search[i].str.replace("<","")
        web_search[i]=pd.to_numeric(web_search[i])
#display information about the content in the csv file
web_search.info()
# show the last 5 elements in dataset
web_search.head()
```

```python
# Convert month to type datetime
pd.to_datetime(web_search["Month"])
```

```python
# display data types and data
web_search.info()
web_search.head()
```

```python
# removing integers as indexing and making months as new index
web_search=web_search.set_index("Month")
web_search.head()
```

```python
#Thus,each sister at some point was more popular than other sisters.
web_search.plot(figsize=(20,10))
```

```python
# Comparing Jenners with Kardashians proved that over the years younger Jenner sisters are more popular than their elder Kardashians sisters.
web_search["Jenners"]=(web_search["Kendall"]+web_search["kylie"])/2
web_search["Kardashians"]=(web_search["kim"]+web_search["khloe"]+web_search["kourtney"])/3
web_search[["Jenners","Kardashians"]].plot()
```
